```python
# Initial imports
import pandas as pd
import numpy as np
import datetime as dt
from pathlib import Path

%matplotlib inline
```

```python
# Reading whale returns
whale_returns_path = Path("./resources/whale_returns.csv")
read_whale_returns = pd.read_csv(
    whale_returns_path, index_col="Date", infer_datetime_format=True, parse_dates=True)
read_whale_returns = read_whale_returns.sort_index()
read_whale_returns.head()
```

```python
# Count nulls
read_whale_returns.isnull().sum()
```

```python
# Drop nulls
read_whale_returns = read_whale_returns.dropna().copy()
read_whale_returns.isnull().sum()
```

```python
# Reading algo returns
algo_returns_path = Path("./resources/algo_returns.csv")
read_algo_returns = pd.read_csv(
    algo_returns_path, index_col="Date", infer_datetime_format=True, parse_dates=True)
read_algo_returns = read_algo_returns.sort_index()
read_algo_returns.head()
```

```python
# Count nulls
read_algo_returns.isnull().sum()
```

```python
# Drop nulls
read_algo_returns = read_algo_returns.dropna().copy()
read_algo_returns.isnull().sum()
```

```python
sp500_history_path = Path("./resources/sp500_history.csv")
read_sp500_history = pd.read_csv(
    sp500_history_path, index_col="Date", infer_datetime_format=True, parse_dates=True)
read_sp500_history = read_sp500_history.sort_index()
read_sp500_history.head()
```

```python
# Rename  the SP500  Column for this portfolio.
read_sp500_history= read_sp500_history.rename(columns={"Close": "S&P 500"},inplace=False)
read_sp500_history.head()
```

```python
read_sp500_history.dtypes
```

```python
read_sp500_history["S&P 500"] = read_sp500_history["S&P 500"].str.replace("$", "")
read_sp500_history["S&P 500"] = read_sp500_history["S&P 500"].astype("float")
read_sp500_history.dtypes
```

```python
# Drop nulls
read_sp500_history = read_sp500_history.dropna().copy()
read_sp500_history.isnull().sum()
```

```python
read_sp500_history.head()
```

```python
read_sp500_returns = read_sp500_history.pct_change()
```

```python
read_sp500_returns.head()
```

```python
# Drop nulls
read_sp500_returns = read_sp500_returns.dropna().copy()
read_sp500_returns.isnull().sum()
```

```python
read_sp500_returns.head()
```

```python
joint_returns = pd.concat([read_whale_returns, read_algo_returns, read_sp500_returns], axis = "columns", join="inner")
joint_returns.head()
```

```python
joint_returns.plot(figsize=(20, 10), title = "Daily Returns")
```

```python
#np.cumprod(1 + df['Daily_rets'].values) - 1
cumulative_portfolio_returns = (1 + joint_returns).cumprod()
cumulative_portfolio_returns.plot(figsize=(20, 10), title = "cumulative returns")
```

```python
print("Berkshire Hathaway and Algo 1 outperfrom S&P 500")
```

```python
joint_returns.plot.box(figsize=(20, 10), title = "returns risk")
```

```python
joint_returns.std()
```

```python
sp500_risk = joint_returns["S&P 500"].std()
joint_returns.std() > sp500_risk
```

```python
annualized_std = joint_returns.std() * np.sqrt(252)
annualized_std
```

```python
joint_returns_roll_std = joint_returns.rolling(window=21).std()
joint_returns_roll_std.plot(figsize=(20, 10), title = "joint_returns_roll_std")
```

```python
returns_correlation = joint_returns.corr() 
returns_correlation
```

```python
returns_correlation.style.background_gradient(cmap="spring")
```

```python
covariance_beta = joint_returns["Algo 1"].rolling(window=60).cov(joint_returns["S&P 500"])
variance_beta = joint_returns["S&P 500"].rolling(60).var()
beta_calc = covariance_beta / variance_beta
beta_calc.plot(figsize=(20, 10), title="Algo 1 Beta Calcualation")
```

```python
joint_returns.ewm(halflife=21).std().plot(figsize=(20, 10), title="Exponentially Weighted Average of ALL Portfolios")
```

```python
sharpe_ratios_all = (joint_returns.mean() * 252) / (joint_returns.std() * np.sqrt(252))
sharpe_ratios_all
```

```python
sharpe_ratios_all.plot(kind="bar", title="Sharpe Ratios All")
```

```python
# Custom porfolio
goog_historical_path = Path("./resources/goog_historical.csv")
read_goog_historical = pd.read_csv(
    goog_historical_path, index_col="Trade DATE", infer_datetime_format=True, parse_dates=True)
read_goog_historical = read_goog_historical.sort_index()
read_goog_historical.head()
```

```python
cost_historical_path = Path("./resources/cost_historical.csv")
read_cost_historical = pd.read_csv(
    cost_historical_path, index_col="Trade DATE", infer_datetime_format=True, parse_dates=True)
read_cost_historical = read_cost_historical.sort_index()
read_cost_historical.head()
```

```python
aapl_historical_path = Path("./resources/aapl_historical.csv")
read_aapl_historical = pd.read_csv(
   aapl_historical_path, index_col="Trade DATE", infer_datetime_format=True, parse_dates=True)
read_aapl_historical = read_aapl_historical.sort_index()
read_aapl_historical.head()
```

```python
joint_stocks = pd.concat([read_goog_historical, read_cost_historical, read_aapl_historical], axis = "rows", join="inner")
joint_stocks.head()
```

```python
#Resets date index
joint_stocks = joint_stocks.reset_index()
joint_stocks.head()
```

```python
joint_stocks
```

```python
# Reorganize portfolio data by having a column per symbol
joint_portfolio = joint_stocks.pivot_table(values="NOCP", index="Trade DATE", columns="Symbol")
joint_portfolio.head()
```

```python
# Calculate daily returns
joint_daily_returns = joint_portfolio.pct_change()

# Drop NAs
joint_daily_returns = joint_daily_returns.dropna().copy()

# Display sample data
joint_daily_returns.head()
```

```python
# Set weights
weights = [1/3, 1/3, 1/3]

# Calculate portfolio return
joint_portfolio_returns = joint_daily_returns.dot(weights)

# Display sample data
joint_portfolio_returns.head()
```

```python
# Join your returns DataFrame to the original returns DataFrame

joint_returns["Custom Portfolio"] = joint_portfolio_returns
joint_returns.tail()
```

```python
# Only compare dates where return data exists for all the stocks (drop NaNs)

joint_returns = joint_returns.dropna().copy()
joint_returns.tail()
```

```python
joint_annualized_std = joint_returns.std() * np.sqrt(252)
joint_annualized_std
```

```python
# Rolling std 21 day window
joint_annualized_roll_std = joint_returns.rolling(window=21).std()
joint_annualized_roll_std.plot(figsize=(20, 10), title = "joint_annualized_roll_std")
```

```python
# Calculate and plot correlation 
joint_returns_correlation = joint_returns.corr() 
joint_returns_correlation
```

```python
joint_returns_correlation.style.background_gradient(cmap="spring")
```

```python
custom_covariance_beta = joint_returns["Custom Portfolio"].rolling(window=60).cov(joint_returns["S&P 500"])
custom_variance_beta = joint_returns["S&P 500"].rolling(60).var()
custom_beta_calc = custom_covariance_beta / custom_variance_beta
custom_beta_calc.plot(figsize=(20, 10), title="Custom Portfolio Beta Calcualation")
```

```python
sharpe_ratios_custom = (joint_returns.mean() * 252) / (joint_returns.std() * np.sqrt(252))
sharpe_ratios_custom
```

```python
sharpe_ratios_custom.plot(kind="bar", title="Sharpe Ratios Custom")
```
